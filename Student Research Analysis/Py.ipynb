# Drop the Unnamed:0

```python
df=df.drop("Unnamed: 0",axis=1)
# axis=1 means vertically 
print(df.head())
```

## Step 1 : Understanding data ( .head()  ,  .describe()  ,  .info()  , .isnull()  ,  .sum() ,  .drop(" ",axis=1 )  )  

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5 - 10") 
# df.head()
print(df)
```

## Gender Distribution 

```python
plt.figure(figsize = (5,5))
ax=sns.countplot(data =df, x ="Gender")
# ax.bar_lable(ax.containers[0])
plt.show()
```

Conclusion : The Female count is more than male 

# Parents Education Distribution

```python
print(df.ParentEduc)
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean" , "ReadingScore":"mean" , "WritingScore":"mean" })
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb , annot=True)
plt.show()
```

## Conclusion : Eduaction of parents impacts the scores of students 

# Gender Distribution

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean" , "ReadingScore":"mean" , "WritingScore":"mean" })
print(gb)
```

```python

plt.figure(figsize=(5,5))
sns.heatmap(gb , annot=True)
plt.title("Gender Distribution ")
plt.show()
```

## Conclusion : Marital status of parents have neglegible impact on students score

# Ethnic Distribution 

```python
print(df["EthnicGroup"].unique())
```

```python
groupA= df.loc[(df['EthnicGroup']) == "group A"].count()
groupB= df.loc[(df['EthnicGroup']) == "group B"].count()
groupC= df.loc[(df['EthnicGroup']) == "group C"].count()
groupD= df.loc[(df['EthnicGroup']) == "group D"].count()
groupE= df.loc[(df['EthnicGroup']) == "group E"].count()
print("Group A : " , groupA["EthnicGroup"])
print("Group B : " , groupB["EthnicGroup"])
print("Group C : " , groupC["EthnicGroup"])
print("Group D : " , groupD["EthnicGroup"])
print("Group E : " , groupE["EthnicGroup"])
```

```python
l=["Group A" ,"Group B" ,"Group C","Group D" ,"Group F" ]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist , labels = l , autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Group")

plt.show()
```
